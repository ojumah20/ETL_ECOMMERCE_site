```python
from bs4 import BeautifulSoup
```

```python
import requests #library to make request to a url
```

```python
#more efficent way using recursion algorithm
url = 'https://www.jumia.com.ng/catalog/?q=phones'
page  = requests.get(url)
soup  =  BeautifulSoup(page.content,'html.parser')
jumia = soup.find_all("div",class_='info')
new_list = []
x = 0

def create_list(jumia,new_list,x):

    if x > len(jumia)-1:
        return new_list
    else:
        name = jumia[x].find("h3",class_ ="name").text
        price = jumia[x].find("div",class_ = "prc").text

        if jumia[x].find("div",class_ = "rev") is not None:
            rating  = jumia[x].find("div",class_ = "rev").text
        else:
            rating  = "not available"

        info = [name,price,rating]
        new_list.append(info)
        return create_list(jumia,new_list,x+1)


create_list(jumia,new_list,x)
```

```python

#using recursion algorithm, faster and more efficient
for j in range (2,50):
    url = f'https://www.jumia.com.ng/catalog/?q=phones&page={j}#catalog-listing'
    page  = requests.get(url)
    soup  =  BeautifulSoup(page.content,'html.parser')
    jumia = soup.find_all("div",class_='info')

    def create_list(jumia,new_list,x):

        if x > len(jumia)-1:
            return new_list
        else:
            name = jumia[x].find("h3",class_ ="name").text
            price = jumia[x].find("div",class_ = "prc").text

            if jumia[x].find("div",class_ = "rev") is not None:
                rating  = jumia[x].find("div",class_ = "rev").text
            else:
                rating  = "not available"

            info = [name,price,rating]
            new_list.append(info)
            return create_list(jumia,new_list,x+1)
            
```
